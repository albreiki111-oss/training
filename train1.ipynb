```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Studentscore.csv')
```

```python
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=200, n_features=2, n_informative=2,
                           n_redundant=0, n_clusters_per_class=1,
                           class_sep=1.5, random_state=42)
```

```python
# Convert to DataFrame
df = pd.DataFrame(X, columns=["gender", "reading score"])
df["Pass"] = y
```

```python
# Display first few rows
print("üìä Sample Data:")
print(df.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df[["gender", "reading score"]], df["Pass"], test_size=0.2, random_state=0)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

```python
def ask_question(index, X_sample, y_sample):
    gender = X_sample.iloc[index, 0]
    reading = X_sample.iloc[index, 1]

    print(f"\n‚ùì Question {index + 1}")
    print(f"gender: {gender:.2f}, reading score: {reading:.2f}")
    user_input = input("Will this person pass the test? (yes/no): ").strip().lower()

    # Model prediction
    sample_scaled = scaler.transform([[gender, reading]])
    prediction = model.predict(sample_scaled)[0]
    correct_answer = 'yes' if prediction == 1 else 'no'

    if user_input == correct_answer:
        print("‚úÖ Correct!")
        return 1
    else:
        print(f"‚ùå Incorrect. The correct answer is {correct_answer}.")
        return 0
```

```python
score = 0
num_questions = 5

print("\nüéØ Starting the Logistic Regression Quiz!")
for i in range(num_questions):
    score += ask_question(i, X_test.reset_index(drop=True), y_test.reset_index(drop=True))

    print(f"\nüèÅ Quiz Finished! Your Score: {score}/{num_questions}")
```
